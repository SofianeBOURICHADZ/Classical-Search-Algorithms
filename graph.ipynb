```python
from collections import deque

romania_map = {
    'Arad': {'Zerind', 'Sibiu', 'Timisoara'},
    'Zerind': {'Arad', 'Oradea'},
    'Oradea': {'Zerind', 'Sibiu'},
    'Sibiu': {'Arad', 'Oradea', 'Fagaras', 'Rimnicu Vilcea'},
    'Timisoara': {'Arad', 'Lugoj'},
    'Lugoj': {'Timisoara', 'Mehadia'},
    'Mehadia': {'Lugoj', 'Drobeta'},
    'Drobeta': {'Mehadia', 'Craiova'},
    'Craiova': {'Drobeta', 'Rimnicu Vilcea', 'Pitesti'},
    'Rimnicu Vilcea': {'Sibiu', 'Craiova', 'Pitesti'},
    'Fagaras': {'Sibiu', 'Bucharest'},
    'Pitesti': {'Rimnicu Vilcea', 'Craiova', 'Bucharest'},
    'Bucharest': {'Fagaras', 'Pitesti', 'Giurgiu', 'Urziceni'},
    'Giurgiu': {'Bucharest'},
    'Urziceni': {'Bucharest', 'Vaslui', 'Hirsova'},
    'Hirsova': {'Urziceni', 'Eforie'},
    'Eforie': {'Hirsova'},
    'Vaslui': {'Urziceni', 'Iasi'},
    'Iasi': {'Vaslui', 'Neamt'},
    'Neamt': {'Iasi'}
}
```

```python
def find_shortest_path_per_node(goal,distances_from_origin):
    start = 'Arad'
    maps = deque()
    maps.append(start)
    arrived = False
    while(len(maps) != -1):
        current_pos = maps[-1]
        maps.pop()
        neighbours = romania_map[current_pos]
        for i in neighbours:
            if distances_from_origin[i] == -1:
                distances_from_origin[i] = distances_from_origin[current_pos] + 1
            if i == goal:
                arrived = True
            maps.appendleft(i)
        if arrived:
            break
    if arrived:
        print(distances_from_origin[goal])
    else:
        print('UNREACHABLE')
```

```python
from heapq import heappush, heappop
def find_sl(goal,distances_from_origin_for_distances):
    start = 'Arad'
    ls = []
    heappush(ls,(distances_from_origin_for_distances[start],start))
    current_pos = heappop(ls)
    while(current_pos[1] != goal):
        neighbours = romania_map_with_distances[current_pos[1]]
        for i in neighbours:
            distances_from_origin_for_distances[i] = current_pos[0] + romania_map_with_distances[current_pos[1]][i]
            heappush(ls,(distances_from_origin_for_distances[i], i))
        current_pos = heappop(ls)
    return current_pos
```
